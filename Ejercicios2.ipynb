```python
numero = input("Ingresa tu número de la suerte: ")

if numero.lstrip("-").isdigit():
    print("Es un número entero.")
    print("Cantidad de dígitos:", len(numero.lstrip("-")))
else:
    print("El valor ingresado no es un número entero.")
```

```python
#Ejercicio 3
```

```python
n1 = int(input("Ingresa el primer número: "))
n2 = int(input("Ingresa el segundo número: "))

if n1 > n2:
    print("El mayor es:", n1)
elif n2 > n1:
    print("El mayor es:", n2)
else:
    print("Ambos números son iguales.")
```

```python
#Ejercicio 4
```

```python
n1 = int(input("Ingresa el primer número: "))
n2 = int(input("Ingresa el segundo número: "))

suma = 0
for i in range(n1, n2 + 1):
    suma += i

print("La suma es:", suma)
```

```python
n1 = int(input("Ingresa el primer número: "))
n2 = int(input("Ingresa el segundo número: "))

inicio = min(n1, n2)
fin = max(n1, n2)

suma = 0
for i in range(inicio, fin + 1):
    suma += i

print("La suma es:", suma)
```
